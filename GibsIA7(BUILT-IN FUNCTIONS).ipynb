# **Reponse**

```python
with open('fichier1.txt', 'w') as f:
  for i in range(10):
    #f.write("{}^2 = {}".format(i, i**2))
    f.write(f"{i}^2 = {i**2}")

with open('fichier1.txt', 'r') as f:
  print(list(f.read()))
```

# **Correction**

```python
with open('fichier2.txt', 'w') as f:
  for i in range(10):
    f.write(f"{i}^2 = {i**2} \n")

with open('fichier2.txt', 'r') as f:
  liste = f.readlines()
liste
```

**Sans le "\n"**

```python
with open('fichier2.txt', 'r') as f:
  liste = f.read().splitlines()
liste
```

**Avec les liste comprehension**

```python
liste = [line for line in open('fichier.txt', 'r')]
liste
```

**Avec les liste comprehension et sans le "\n**

```python
liste = [line.strip() for line in open('fichier.txt', 'r')]
liste
```
