```python
def e_potentielle_mars(masse, hauteur, e_limite, g):
    E = masse*hauteur*g
    if(E<=e_limite):
      print(E,'Energie acceptable')
    else:
      print('Energie inacceptable')

e_potentielle_mars(80, 8, 10000, 9.81)
```
