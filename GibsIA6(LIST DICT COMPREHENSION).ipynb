# **Tous les carres de 0 a 9 **

```python
import time
import numpy as np
```

```python
start = time.time()

liste_1 = []
for i in range(10000000):
  liste_1.append(i**2)

end = time.time()
#print(liste_1)
print(end-start)
```

```python
start = time.time()

liste_2 = [i**2 for i in range(10000000)] #--> liste comprehension

end = time.time()
#print(liste_2)
print(end-start)
```

```python
liste_3 = [[i for i in range(3)] for j in range(5)]
liste_3
```

# **Dictionnaire comprehension**

```python
dictionnaire = {
    '0': 'Pierre',
    '1': 'Jean',
    '2': 'Julie',
    '3': 'Sophie'
}
dictionnaire
```

```python
prenoms = ['Pierre','Jean','Julie','Sophie']
```

```python
dico = {k:v for k, v in enumerate(prenoms)}
dico
```

```python
dico.keys()
```

```python
dico.values()
```

```python
ages = [24, 62, 10, 23]
```

```python
dico_2 = {prenoms:ages for prenoms, ages in zip(prenoms,ages)}
dico_2
```

```python
dico_2 = {prenoms:ages for prenoms, ages in zip(prenoms,ages) if ages > 20}
dico_2
```

# **Tuple**

```python
tuple_1 = (i**2 for i in range(10))
tuple_1
```

```python
tuple_1 = tuple((i**2 for i in range(10)))
tuple_1
```
