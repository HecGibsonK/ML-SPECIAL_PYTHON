```python
import numpy as np
```

# **Fonction**

```python
x = np.linspace(0, 2, 10)
y = x**2
print(x)
```

# 1) **Pyploy**

```python
import matplotlib.pyplot as plt
```

```python
plt.figure() # Debut de la figure;   plt.figure(figsize=(12, 8))-Pour donner une dimenssion a notre figure

plt.plot(x, y, label='Quadratique', c='black', lw=2, ls='-') # label=..., lw=..., ls=..., c=...
plt.plot(x, x**3, label='Cubique', c='red')
plt.title('Figure 1')
plt.xlabel('axe x')
plt.ylabel('axe y')
plt.legend()

plt.show()  # Obligatoire si ce n'est pas un editeur "Jupiter"  /// Affiche la figure

plt.savefig('text.png')

plt.figure() # Si vous n'etes pas sur "Jupiter"
plt.plot(x, y)
```

```python
plt.subplot(2, 1, 1)
plt.plot(x, y, label='Graphique 1')
plt.subplot(2, 1, 2)
plt.plot(x, x**3, label='Graphique 2', c='red')
```

# **Matplotlib Oriente Objet**

```python
fig, ax = plt.subplots()
ax.plot(x,y)
```

```python
fig, ax = plt.subplots(2, 1, sharex=True)
ax[0].plot(x,y)
ax[1].plot(x,np.sin(x))
```

# **Correction**

```python
datasets1 = {
    f"experience{i}": np.random.randn(100, 3)
    for i in range(3)
}
```

```python
def graphiqu1(datasets1):
  n = len(datasets1)
  plt.figure(figsize=(12,15))
  for k, i in zip(datasets1.keys(), range(1, n+1)):
    plt.subplot(n,1,i)
    plt.plot(datasets1[k])
    plt.title(k)
  plt.show()
graphiqu1(datasets1)
```
