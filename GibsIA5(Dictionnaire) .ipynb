# **Exercise**

```python
classeur = {
    "positif" : [],  #[5,4,6,2,8,7,6,4,9,5],
    "negatif" : [],  #[-9,-5,-6,-1,-45,-78,-34,-96,-84,-4],
}
```

```python
def trier(classeur, nombre):
  # classeur: dictionnaire taille 2
  # nombre: int
  # Range nombre dans "positif" ou "negatif" de classeur selon le signe de nombre
  if (nombre>=0):
    classeur["positif"] = nombre
  else:
    classeur["negatif"] = nombre
  return classeur
```

```python
nombre = -8
trier(classeur, nombre)
```

```python
classeur1 = {
    "positif" : [],  #[5,4,6,2,8,7,6,4,9,5],
    "negatif" : [],  #[-9,-5,-6,-1,-45,-78,-34,-96,-84,-4],
}
```

```python
def trier_C(classeur1, nombre):
  # classeur: dictionnaire taille 2
  # nombre: int
  # Range nombre dans "positif" ou "negatif" de classeur selon le signe de nombre
  if (nombre>=0):
    classeur1["positif"].append(nombre)
  else:
    classeur1["negatif"].append(nombre)
  return classeur1
trier_C(classeur1, nombre)
```
