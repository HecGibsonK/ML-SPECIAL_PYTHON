# **Exercice**

```python
#np.random.seed(0)
def initialisation(m,n):
  # m: nombre de lignes
  # n: nombre de colonnes
  # retourne une matrice aleatoire (m, n+1)
  # avec une colonne biais (remplie de "1") tout a droite

  MA1 = np.random.randn(m,n)

  MA2 = np.ones((m,1))

  MA = np.concatenate((MA1,MA2), axis=1)

  return MA
```

```python
m = int(input('Entrez la valeur de m \n'))
n = int(input('Entrez la valeur de n \n'))
#m = 5
#n = 4
initialisation(m,n)
```

# **Correction**

```python
def initialisation1(m,n):
  # m: nombre de lignes
  # n: nombre de colonnes
  # retourne une matrice aleatoire (m, n+1)
  # avec une colonne biais (remplie de "1") tout a droite

  X = np.random.randn(m,n)
  X = np.concatenate((X, np.ones((X.shape[0],1))), axis=1)

  return X
```

```python
m = int(input('Entrez la valeur de m \n'))
n = int(input('Entrez la valeur de n \n'))
#m = 5
#n = 4
initialisation1(m,n)
```
