# **If/Elif/Else --- for and While**

```python
x= 0
if x > 0:
  print(x,'positif')
elif x==0:
  print(x,'nul')
else:
  print(x,'negatif')
```

```python
def signe(x):
  if(x>0):
    print(x,'positif')
  elif x==0:
    print(x,'nul')
  else:
    print(x,'negatif')
```

# **Exercice**
Retourne la Suite de Fibonacci a 'n' valeur

# **Ma reponse**

```python
def fibonacci(n):
  SF = np.arange(start=0,stop=n)
  for i in range(n):
    SF[0] = 0
    SF[1] = 1
    SF[i] = SF[i-1] + SF[i-2]
  return SF
fibonacci(11)
```
